```python
import os
import cv2
import numpy as np

def add_gaussian_noise(input_folder, output_folder, mean=0, std=0.5):
    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Get a list of all files in the input folder
    files = os.listdir(input_folder)

    # Process each file in the input folder
    for file in files:
        # Read the image
        image_path = os.path.join(input_folder, file)
        image = cv2.imread(image_path)

        # Add Gaussian noise
        noise = np.random.normal(mean, std, image.shape).astype(np.uint8)
        noisy_image = cv2.add(image, noise)

        # Save the modified image to the output folder
        output_path = os.path.join(output_folder, file)
        cv2.imwrite(output_path, noisy_image)

        print(f"Processed: {file}")

# Specify the input and output folder paths
input_folder = 'FoETrainingSets180'
output_folder = 'training_images'

# Call the function to add Gaussian noise to the images
add_gaussian_noise(input_folder, output_folder)
```
